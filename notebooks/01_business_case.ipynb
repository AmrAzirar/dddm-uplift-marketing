## 1. Contexte Métier

Un retailer américain spécialisé dans la vente en ligne a mené une campagne 
email promotionnelle auprès d'une partie de sa base client de 64 000 personnes.

Le problème classique du marketing digital :
> *"On envoie des emails à tout le monde — mais combien de ces achats 
> auraient eu lieu sans l'email ?"*

### Les 4 types de clients qu'on cherche à identifier :

| Segment | Comportement | Action recommandée |
|---|---|---|
| 🎯 **Persuadables** | Achètent GRÂCE à l'email | Cibler en priorité |
| ✅ **Sure Things** | Auraient acheté SANS email | Ne pas contacter = économie |
| 😴 **Lost Causes** | N'achètent pas dans tous les cas | Ignorer |
| 🚫 **Sleeping Dogs** | Se désintéressent quand contactés | Ne SURTOUT pas contacter |

## 2. Question Décisionnelle Centrale

> **"Parmi les 64 000 clients du retailer, quels sont ceux pour lesquels 
> l'envoi d'un email promotionnel génère un achat incrémental réel — 
> afin de maximiser le ROI de la campagne en ciblant uniquement 
> les Persuadables et en évitant les Sure Things et Sleeping Dogs ?"**

### Pourquoi cette question est différente :

- ❌ Question classique : *"Ce client va-t-il acheter ?"* → Prédiction
- ✅ Notre question : *"Est-ce que l'email CAUSE l'achat ?"* → Causalité

Cette distinction est fondamentale : un modèle prédictif classique 
ciblerait les clients les plus susceptibles d'acheter — ce qui inclut 
les Sure Things (budget gaspillé). Notre approche par **Uplift Modeling** 
mesure l'effet CAUSAL de l'email sur chaque client.

## 3. KPI Tree — De l'Objectif Stratégique aux Métriques

### Objectif Stratégique
**Maximiser le ROI de la campagne email en ciblant intelligemment**

### Arbre des KPIs

- 📈 **KPIs Primaires** (résultats business)
    - Uplift moyen global (conversion incrémentale)
    - ROI campagne ciblée vs campagne blast
    - Budget économisé (en évitant Sure Things)

- 🎯 **KPIs Secondaires** (performance modèle)
    - AUUC — Area Under Uplift Curve
    - Qini Coefficient
    - % clients correctement segmentés

- 📊 **Métriques Opérationnelles** (données brutes)
    - Taux de conversion par segment (visit, conversion)
    - Dépense moyenne par groupe (spend)
    - Taux de réponse email homme vs femme
    - Distribution des 4 personas clients

### Définition des KPIs Primaires

| KPI | Formule | Objectif Cible |
|---|---|---|
| **Uplift moyen** | Conv(traitement) - Conv(contrôle) | > 2% |
| **ROI ciblé** | Revenus incrém. / Coût campagne ciblée | > 150% |
| **Budget économisé** | Coût emails Sure Things évités | Maximiser |
| **AUUC** | Aire sous la courbe uplift | > 0.55 |

## 4. Business Case — Estimation du ROI

### Hypothèses de base

| Paramètre | Valeur | Source |
|---|---|---|
| Nombre total de clients | 64 000 | Dataset Hillstrom |
| Coût d'envoi par email | 0.50 $ | Standard industrie |
| Panier moyen | 100 $ | Dataset (colonne spend) |
| Taux de conversion moyen | 9% | À calculer Phase 3 |
| % estimé Sure Things | ~30% | Hypothèse initiale |

### Scénario 1 — Campagne Blast (sans uplift modeling)
- Emails envoyés : 42 694 (groupes traitement)
- Coût total : 42 694 × 0.50$ = **21 347 $**
- Conversions réelles dues à l'email : estimées ~5%
- Budget gaspillé sur Sure Things : ~30% = **6 404 $**

### Scénario 2 — Campagne Ciblée (avec uplift modeling)
- On cible uniquement les Persuadables (~40% de la base)
- Emails envoyés : ~17 000
- Coût total : 17 000 × 0.50$ = **8 500 $**
- Même nombre de conversions incrémentales
- **Économie réalisée : ~12 847 $**

### ROI Attendu de la Démarche Data
- Investissement modélisation : ~200h travail
- Économie par campagne : ~12 000$
- **ROI = [(12 000 - 0) / 200h] = 60$/heure de travail analytique**

> 💡 Ces chiffres seront affinés et validés en Phase 3 
> avec les vraies données du dataset.
