# A/B Test Plan — Campagne Email Uplift Marketing
**Projet :** DDDM — Uplift Marketing Hillstrom  
**Auteur :** Azirar Amr / Anas Houhou 
**Date :** Juin 2026  
**Module :** Data-Driven Decision Making — ENSIAS

---

## 1. Contexte et Objectif

Le modèle uplift développé sur le dataset Hillstrom identifie
15 999 clients Persuadables sur 64 000. L objectif de ce test
est de valider en conditions réelles que cibler uniquement
les Persuadables génère un meilleur ROI que la campagne blast.

---

## 2. Hypothèses

**H0 (nulle) :**
La campagne ciblée (Persuadables uniquement) ne génère pas
de revenu incrémental supérieur à la campagne blast.

**H1 (alternative) :**
La campagne ciblée génère un revenu incrémental
significativement supérieur à la campagne blast.

---

## 3. Design Expérimental

| Paramètre | Valeur |
|---|---|
| Groupe A (contrôle) | Campagne blast — tous les clients |
| Groupe B (traitement) | Campagne ciblée — Persuadables uniquement |
| Randomisation | Assignation aléatoire au niveau client |
| Durée | 4 semaines |
| Période | Semaine 1 à 4 du mois suivant |

### Règles d assignation
- Chaque client est assigné une seule fois
- Assignation aléatoire stratifiée par zone géographique
- Les Sleeping Dogs sont exclus des deux groupes

---

## 4. Calcul de la Taille d Échantillon

| Paramètre | Valeur | Justification |
|---|---|---|
| Alpha | 0.05 | Seuil standard 5% |
| Puissance | 80% | Standard industrie |
| Taux baseline | 0.57% | Groupe No Email Hillstrom |
| Taux attendu | 1.25% | Groupe Mens Email Hillstrom |
| Effect size | 0.68% | Uplift observé |
| Taille par groupe | 3 061 | Calculé scipy.stats |
| Taille totale | 6 122 | 2 x 3 061 |
| Base disponible | 64 000 | Dataset Hillstrom |
| Faisabilité | OUI | 10x au-dessus du minimum |

---

## 5. Métriques de Suivi

### Métrique Primaire
- **Taux de conversion incrémental**
  Formule : Conv(traitement) - Conv(contrôle)
  Objectif : > 0.68%

### Métriques Secondaires
- Revenu incrémental par email envoyé
- ROI campagne (%)
- Taux de désabonnement
- Taux de visite site web

### Métriques de Guardrail
- Taux de désabonnement < 2%
- Taux de plainte spam < 0.1%

---

## 6. Critères de Décision

### Succès
- P-value < 0.05 après 4 semaines
- Uplift observé > 0.50%
- Taux désabonnement < 2%

### Arrêt anticipé
- P-value < 0.01 après 2 semaines
- Taux désabonnement > 2%
- Taux plainte spam > 0.1%

### Échec
- P-value > 0.05 après 4 semaines
- Uplift observé < 0.30%

---

## 7. Impact Financier Attendu

| Scénario | Emails | Coût | ROI Incrémental |
|---|---|---|---|
| Blast (A) | 42 614 | 21 307$ | 193% |
| Ciblée (B) | 15 999 | 8 000$ | 328% |
| Économie | -26 615 | -13 308$ | +70% |

**Conclusion :** Pour 1$ de profit incrémental sacrifié,
on économise 9$ de budget campagne.
